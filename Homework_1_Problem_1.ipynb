```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("test.txt") 
letters = []
lettersDict = {}
lettersNumpy = []
while 1:
    char = f.read(1)    
    if char.isalnum():
        letters.append(char.lower())
    if not char:
        break
f.close()
letters.sort()
index = -1
for char in letters:
    if char in lettersDict:
        lettersDict[char] += 1
    else:
        lettersDict[char] = 0
for key in lettersDict:
    lettersNumpy.append(lettersDict[key])
lettersNumpy = np.array(lettersNumpy, dtype="float")
print(lettersDict)
```

```python
plt.bar(lettersDict.keys(), lettersDict.values())
plt.title('Unnormalized')
plt.xlabel('Letters')
plt.ylabel('Frequency')
plt.show()
```

```python
normNum=np.linalg.norm(lettersNumpy)
norm = lettersNumpy/normNum
plt.bar(lettersDict.keys(), norm)
plt.title('Normalized')
plt.xlabel('Letters')
plt.ylabel('Frequency')
plt.show()
```
